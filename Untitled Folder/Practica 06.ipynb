```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
import scipy.io as sio 
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score
import math
from process_email import *
from get_vocab_dict import *
import codecs
```

```python
data = loadmat("ex6data1.mat")
data_2 = loadmat("ex6data2.mat")
data_3 = loadmat("ex6data3.mat")
X = data["X"]
Y = data["y"]
X_2 = data_2["X"]
Y_2 = data_2["y"]
X_3 = data_3["X"]
Y_3 = data_3["y"]
Xval_3=data_3["Xval"]
Yval_3=data_3["yval"]
```

```python
# Obtiene un vector con los índices  de  los  ejemplos  positivos
pos_pass = np.where(Y == 1)
pos_fail = np.where(Y == 0)
```

```python
#Dibuja los ejemplos positivos
plt.scatter(X[pos_pass, 0], X[pos_pass, 1], color='black', marker='+')
plt.scatter(X[pos_fail, 0], X[pos_fail, 1], color='yellow', edgecolors='black', marker='o')
plt.title("Conjunto de datos 1")
```

```python
def visualize_boundary(X, y, svm):
    x1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
    x2 = np.linspace(X[:, 1].min(), X[:, 1].max(), 100)
    x1, x2 = np.meshgrid(x1, x2)
    yp = svm.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape)
    pos = (y == 1).ravel()
    neg = (y == 0).ravel()
    plt.figure()
    plt.scatter(X[pos, 0], X[pos, 1], color='black', marker='+')
    plt.scatter(X[neg, 0], X[neg, 1], color='yellow', edgecolors='black', marker='o')
    plt.contour(x1, x2, yp)
    plt.show()
    plt.close()
```

```python
svm = SVC(kernel='linear', C=1)
svm.fit(X, Y.ravel())
visualize_boundary(X,Y,svm)
```

```python
svm = SVC(kernel='linear', C=100)
svm.fit(X, Y.ravel())
visualize_boundary(X,Y,svm)
```

### Kernel gaussiano

```python
# Obtiene un vector con los índices  de  los  ejemplos  positivos
pos_pass = np.where(Y_2 == 1)
pos_fail = np.where(Y_2 == 0)
```

```python
#Dibuja los ejemplos positivos
plt.scatter(X_2[pos_pass, 0], X_2[pos_pass, 1], color='black', marker='+')
plt.scatter(X_2[pos_fail, 0], X_2[pos_fail, 1], color='yellow', edgecolors='black', marker='o')
plt.title("Conjunto de datos 2")
```

```python
sigma = 0.1
svm = SVC(kernel='rbf', C=1, gamma=1/(2 * sigma**2))
svm.fit(X_2, Y_2.ravel())
visualize_boundary(X_2,Y_2,svm)
```

### Elección de parámetros C y  σ

```python
def train_with_params(X,Y,Xval,Yval):
    params = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    best_svm = None
    best_accuracy = 0
    for C in params:
        for sigma in params:
            svm = SVC(kernel='rbf', C=C, gamma=1/(2 * sigma**2))
            svm.fit(X, Y.ravel())
            acc = accuracy_score(Yval, svm.predict(Xval))
            if(best_accuracy < acc ):
                best_svm = svm
                best_accuracy = acc

    return best_svm, best_accuracy

visualize_boundary(X_3,Y_3,train_with_params(X_3,Y_3,Xval_3,Yval_3)[0])
        
```

```python
email_contens = open('spam/0001.txt', 'r').read()
email = email2TokenList(email_contens)
```

```python
vocab_dict = getVocabDict()
```

```python
def get_X(email_content):
    email = email2TokenList(email_contens)
    X=np.zeros(len(vocab_dict.keys()))
    for key in vocab_dict.keys():
        if (key in email):
            X[vocab_dict[key]] = 1
    return X
```

```python
def read_email_directory(directory, size):
    emails = []
    for i in range(1,size + 1):
        email_contents = codecs.open('{0}/{1:04d}.txt'.format(directory, i), 'r',encoding='utf-8', errors='ignore').read()
        emails.append(get_X(email_contens))
    return emails
```

```python
X_spam = read_email_directory("spam", 500)
Y_spam = np.ones(500)
X_easy_ham = read_email_directory("easy_ham", 2551)
Y_easy_ham = np.zeros(2551)
X_hard_ham = read_email_directory("hard_ham", 250)
Y_hard_ham = np.zeros(250)
```

```python
X_email = np.concatenate((X_spam,X_easy_ham,X_hard_ham))
Y_email = np.concatenate((Y_spam,Y_easy_ham,Y_hard_ham))
```

```python
print("La mejor precisión obtenida es {}".format(train_with_params(X_email,Y_email,X_email,Y_email)[1]))
```
